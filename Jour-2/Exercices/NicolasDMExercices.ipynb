**Exercice 2 :** Mapping

```python
import pandas as pd
data={'Name':['A','B','C','D'],'Age':[42,54,20,63]}
df=pd.DataFrame(data,index=['value1','value2','value3','value4'])
df
```

```python
score={'A':24,'B':30,'C':36,'D':42}
# Conversion du premier caractère du nom en majuscule pour que les noms soient identiques. 
n=df["Name"].str.capitalize()
n
```

Ecrire un programme Pandas qui sert à ajouter la variable de score à l'ensemble de données df à l'aide de la méthode map(). 

```python
df['Score'] = n.map(score)
df
```

**Exercice 4 :** Renommer les colonnes

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(np.arange(12).reshape(3,4),index=[0,1,2],columns=["a","b","c","d"])
df
```

```python
#Create a variable and pass it into the dataset
s=pd.Series(["one","two","three"])
df.index=df.index.map(s)
df
```

Mettez en majuscule les noms de ligne et de colonne avec la méthode de renommage 

![image](https://user-images.githubusercontent.com/123748165/226334330-bbd36d49-8a54-4a71-a734-9db99c0e79ae.png)

```python
df.index = s.str.capitalize()
df.columns = df.columns.str.capitalize()
df
```
